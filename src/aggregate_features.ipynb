```python
import pandas as pd 
```

```python
DATA_DIR = '../data'
BASE_DATASET = f'{DATA_DIR}/base_dataset.csv'

NEAREST_LAKES = f'{DATA_DIR}/nearest_lakes.csv'
ALTITUDE = f'{DATA_DIR}/fixed_elevation_dataset.csv'
WEATHER_STATS = f'{DATA_DIR}/hpp-database-with-weather-stats.csv'
```

```python
base_df = pd.read_csv(BASE_DATASET)
with_attributes = base_df.rename(columns={
    'id': 'plant_id'
})
```

```python
altitude_df = pd.read_csv(ALTITUDE)[[
    'id', 
    'altitude (m)'
]]
altitude_df.rename(columns={
    'id': 'plant_id',
    'altitude (m)': 'altitude_m'
}, inplace = True)
with_attributes = with_attributes.merge(altitude_df, left_on='plant_id', right_on='plant_id')
```

```python
lakes_df = pd.read_csv(NEAREST_LAKES)[[
    'plant_id',
    'distance_km'
]]
lakes_df.rename(columns={
    'distance_km': 'nearest_lake_dist_km'
}, inplace = True)
with_attributes = with_attributes.merge(lakes_df, left_on='plant_id', right_on='plant_id')
```

```python
weather_df = pd.read_csv(WEATHER_STATS)[[
    'id',
    'Days of Rain',
    'Inches of Rain',
    'Avg High Temp',
    'Avg Low Temp'
]]
weather_df.rename(columns={
    'id': 'plant_id',
    'Days of Rain': 'days_of_rain',
    'Inches of Rain': 'inches_of_rain',
    'Avg High Temp': 'avg_high_temp',
    'Avg Low Temp': 'avg_low_temp'
}, inplace = True)
with_attributes = with_attributes.merge(weather_df, left_on='plant_id', right_on='plant_id')
```
