```python
import pandas as pd 
```

```python
DATA_DIR = '../../data'
BASE_DATASET = f'{DATA_DIR}/base_dataset.csv'

NEAREST_LAKES = f'{DATA_DIR}/nearest_lakes.csv'
ALTITUDE = f'{DATA_DIR}/fixed_elevation_dataset.csv'
WEATHER_STATS = f'{DATA_DIR}/hpp-database-with-weather-stats.csv'
WEATHER_EOBS = f'{DATA_DIR}/hpp-database-with-EOBS-climate.csv'
GRADIENT = f'{DATA_DIR}/BaseDatasetWithGradient.csv'

OUTPUT = f'{DATA_DIR}/hydropower_efficiency.csv'
```

```python
base_df = pd.read_csv(BASE_DATASET)
with_attributes = base_df.rename(columns={
    'id': 'plant_id'
})
```

```python
altitude_df = pd.read_csv(ALTITUDE)[[
    'id', 
    'altitude (m)'
]]
altitude_df.rename(columns={
    'id': 'plant_id',
    'altitude (m)': 'altitude_m'
}, inplace = True)
with_attributes = with_attributes.merge(altitude_df, left_on='plant_id', right_on='plant_id')
```

```python
lakes_df = pd.read_csv(NEAREST_LAKES)[[
    'plant_id',
    'distance_km'
]]
lakes_df.rename(columns={
    'distance_km': 'nearest_lake_dist_km'
}, inplace = True)
with_attributes = with_attributes.merge(lakes_df, left_on='plant_id', right_on='plant_id')
```

```python
weather_df = pd.read_csv(WEATHER_STATS)[[
    'id',
    'Days of Rain'
]]
weather_df.rename(columns={
    'id': 'plant_id',
    'Days of Rain': 'days_of_rain'
}, inplace = True)
with_attributes = with_attributes.merge(weather_df, left_on='plant_id', right_on='plant_id')
```

```python
weather_eobs_df = pd.read_csv(WEATHER_EOBS)[[
    'id',
    'avg daily temp',
    'min daily temp',
    'max daily temp',
    'sea level pressure',
    'global radiation',
    'rainfall'
]]
weather_eobs_df.rename(columns={
    'id': 'plant_id',
    'avg daily temp': 'avg_daily_temp',
    'min daily temp': 'min_daily_temp',
    'max daily temp': 'max_daily_temp',
    'sea level pressure': 'sea_level_pressure',
    'global radiation': 'global_radiation'
}, inplace = True)
with_attributes = with_attributes.merge(weather_eobs_df, left_on='plant_id', right_on='plant_id')
```

```python
gradient_df = pd.read_csv(GRADIENT)[[
    'id',
    '50m Gradient',
    '100m Gradient',
    '500m Gradient'
]]
gradient_df.rename(columns={
    'id': 'plant_id',
    '50m Gradient': '50m_gradient',
    '100m Gradient': '100m_gradient',
    '500m Gradient': '500m_gradient'
}, inplace = True)
with_attributes = with_attributes.merge(gradient_df, left_on='plant_id', right_on='plant_id')
```

```python
with_attributes.to_csv(OUTPUT, index=False)
```
