### Add columns to base dataset

```python
import pandas as pd
import seaborn as sns
```

```python
base_dataset = pd.read_csv('../data/base_dataset.csv')
df_w_weather = base_dataset
df_w_weather['rainfall'] = 0.0
df_w_weather['avg daily temp'] = 0.0
df_w_weather['min daily temp'] = 0.0
df_w_weather['max daily temp'] = 0.0
df_w_weather['sea level pressure'] = 0.0
df_w_weather['global radiation'] = 0.0
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

```python
#tg

tg_list = []
for i in range(df_w_weather.shape[0]):
    lat = df_w_weather['latitude'][i]
    lon = df_w_weather['longitude'][i]
    lat_ind = find_nearest(tg_lat, lat)
    lon_ind = find_nearest(tg_lon, lon)
    avg_tg = 0
    for i in range(tg_time.shape[0]):
        val = tg_avg_temp[i, lat_ind, lon_ind]
        if val > -9999:
            avg_tg += tg_avg_temp[i, lat_ind, lon_ind]
    avg_tg /= tg_time.shape[0]
    tg_list.append(avg_tg)
df_w_weather['avg daily temp'] = tg_list
```

```python
#tn

tn_list = []
for i in range(df_w_weather.shape[0]):
    lat = df_w_weather['latitude'][i]
    lon = df_w_weather['longitude'][i]
    lat_ind = find_nearest(tn_lat, lat)
    lon_ind = find_nearest(tn_lon, lon)
    avg_tn = 0
    for i in range(tn_time.shape[0]):
        val = tn_min_temp[i, lat_ind, lon_ind]
        if val > -9999:
            avg_tn += tn_min_temp[i, lat_ind, lon_ind]
    avg_tn /= tn_time.shape[0]
    tn_list.append(avg_tn)
df_w_weather['min daily temp'] = tn_list
```

```python
#tx

tx_list = []
for i in range(df_w_weather.shape[0]):
    lat = df_w_weather['latitude'][i]
    lon = df_w_weather['longitude'][i]
    lat_ind = find_nearest(tx_lat, lat)
    lon_ind = find_nearest(tx_lon, lon)
    avg_tx = 0
    for i in range(tx_time.shape[0]):
        val = tx_max_temp[i, lat_ind, lon_ind]
        if val > -9999:
            avg_tx += tx_max_temp[i, lat_ind, lon_ind]
    avg_tx /= tx_time.shape[0]
    tx_list.append(avg_tx)
df_w_weather['max daily temp'] = tx_list
```

```python
#rr

rainfall_list = []
for i in range(df_w_weather.shape[0]):
    lat = df_w_weather['latitude'][i]
    lon = df_w_weather['longitude'][i]
    lat_ind = find_nearest(rr_lat, lat)
    lon_ind = find_nearest(rr_lon, lon)
    avg_rr = 0
    for i in range(rr_time.shape[0]):
        val = rr_rainfall[i, lat_ind, lon_ind]
        if val > -9999:
            avg_rr += rr_rainfall[i, lat_ind, lon_ind]
    avg_rr /= rr_time.shape[0]
    rainfall_list.append(avg_rr)
df_w_weather['rainfall'] = rainfall_list
```

```python
#pp

pp_list = []
for i in range(df_w_weather.shape[0]):
    lat = df_w_weather['latitude'][i]
    lon = df_w_weather['longitude'][i]
    lat_ind = find_nearest(pp_lat, lat)
    lon_ind = find_nearest(pp_lon, lon)
    avg_pp = 0
    for i in range(pp_time.shape[0]):
        val = pp_pressure[i, lat_ind, lon_ind]
        if val > -9999:
            avg_pp += val
    avg_pp /= pp_time.shape[0]
    pp_list.append(avg_pp)
df_w_weather['sea level pressure'] = pp_list
```

```python
#qq

qq_list = []
for i in range(df_w_weather.shape[0]):
    lat = df_w_weather['latitude'][i]
    lon = df_w_weather['longitude'][i]
    lat_ind = find_nearest(qq_lat, lat)
    lon_ind = find_nearest(qq_lon, lon)
    avg_qq = 0
    for i in range(pp_time.shape[0]):
        val = qq_radiation[i, lat_ind, lon_ind]
        if val > -9999:
            avg_qq += qq_radiation[i, lat_ind, lon_ind]
    avg_qq /= qq_time.shape[0]
    qq_list.append(avg_qq)
df_w_weather['global radiation'] = qq_list
```
